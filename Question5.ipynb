```python
from scipy.optimize import bisect
import numpy as np


def f(x):
    return (func_d(x)-370.4)

a, b = 0.1,1

root = bisect(f, a, b)

value_at_root = f(root)

print(f"Root: {root}")
print(f"Value at Root: {value_at_root}")
```
